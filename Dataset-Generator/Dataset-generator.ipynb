```python
import requests
import bs4
from bs4 import BeautifulSoup
import pandas as pd
import time
import csv
```

```python
def extract_job_postings(soup):
    div=soup.findAll("div",{"data-tn-component":"organicJob"})
    for i in range(len(div)):
        job_posting=[]
        try:
            job_posting.append(div[i].find("h2").find("a")["title"])
            job_posting.append(div[i].find("span",{"class":"company"}).text.strip())
            job_posting.append(div[i].find("span",{"class":"location"}).text.strip())
            try:
                job_posting.append(div[i].find("span",{"class":"summary"}).text.strip())
            except:
                job_posting.append("NOT-FOUND")
            print("Success")
            return job_posting
        except:
            print("NOT-FOUND")
    return None
```

```python
payload={
    "India":{
        "link":"https://www.indeed.co.in",
        "cities":["Bangalore","Delhi","Mumbai","Hyderabad","Pune"]
    },
    "Singapore":{
        "link":"https://www.indeed.com.sg",
        "cities":["Singapore"]
    },
    "Switzerland":{
        "link":"https://www.indeed.ch",
        "cities":["Zurich"]
    },
    "UK":{
        "link":"https://www.indeed.co.uk",
        "cities":["Cambridge","Manchester","London"]
    },
    "USA":{
        "link":"https://www.indeed.com",
        "cities":["New+York","Chicago","San+Francisco", "Austin", "Seattle", "Los+Angeles", "Philadelphia", "Atlanta", "Dallas", "Pittsburgh", "Portland", "Phoenix", "Denver", "Houston", "Miami", "Washington+DC", "Boulder"]
    }
}
```

```python
jobs=["Software+Developer","Machine+Learning+Engineer","Data+Scientist","Business+Analyst","Mobile+Developer"]
countries=["India","Singapore","Switzerland","UK","USA"]
URL="https://www.indeed.co.in/jobs?q=software+engineer&l=Bangalore&start=100"

file=open('job_postings.csv', 'a')
newFileWriter = csv.writer(file)
newFileWriter.writerow(["job_title", "company","location","summary"])

for country in countries:
    for city in payload[country]["cities"]:
        for job in jobs:
            for i in range(0,1000,10):
                URL=payload[country]["link"]+"/jobs?q="+job+"&l="+city+"&start="+str(i)
                page = requests.get(URL)
                soup = BeautifulSoup(page.text, "html.parser")
                job_posting=extract_job_postings(soup)
                print(job_posting)
                if job_posting is not None:
                    newFileWriter.writerow(job_posting)
```

```python
file.close()
```
